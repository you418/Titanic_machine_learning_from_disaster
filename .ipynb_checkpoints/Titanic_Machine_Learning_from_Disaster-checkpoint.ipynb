## II Feature Engineering

First of all, the label is removed from the traning dataset and then combined with the testing dataset.

```python
#Combine the traning and testing data
def get_combined():
    # extracting and then removing the targets from the training data
    targets = df_train.Survived
    df_train.drop(['Survived'], axis = 1, inplace = True)
    
    combined = df_train.append(df_test)
    combined.reset_index(inplace=True)
    combined.drop(['index', 'PassengerId'], inplace=True, axis=1)
    return combined
```

```python
df_combine = get_combined()
```

### Process Name and Title

In this part, the title is extracted from the name column and put into a new colume "Title".

```python
df_combine['Title'] = df_combine.Name.apply(lambda name:name.split(',')[1].split('.')[0].strip())
```

```python
Title_Dictionary = {
    "Capt": "Officer",
    "Col": "Officer",
    "Major": "Officer",
    "Jonkheer": "Royalty",
    "Don": "Royalty",
    "Sir" : "Royalty",
    "Dr": "Officer",
    "Rev": "Officer",
    "the Countess":"Royalty",
    "Mme": "Mrs",
    "Mlle": "Miss",
    "Ms": "Mrs",
    "Mr" : "Mr",
    "Mrs" : "Mrs",
    "Miss" : "Miss",
    "Master" : "Master",
    "Lady" : "Royalty"
}

def get_titles():
    # we extract the title from each name
    df_combine['Title'] = df_combine['Name'].map(lambda name:name.split(',')[1].split('.')[0].strip())
    
    # a map of more aggregated title
    # we map each title
    df_combine['Title'] = df_combine.Title.map(Title_Dictionary)
    return df_combine
```

```python
df_combine = get_titles()
```

```python
df_combine.Title.value_counts()
```

### Process name and titles

In this section, the orgninal name column is dropped. The 'Title' is changed to dummy variables with the prefix "Title".

```python
def process_names():
    global df_combine
    df_combine.drop('Name', axis=1, inplace=True)
    
    # encoding in dummy variable
    titles_dummies = pd.get_dummies(df_combine['Title'], prefix='Title')
    df_combine = pd.concat([df_combine, titles_dummies], axis=1)
    
    return df_combine

df_combine = process_names()
```

### Processing age

As the column Age has a lot of missing values. The missing values are filled with the median age in the corresponding sex, Pclass, title groups.

```python
df_combine_train = df_combine.iloc[:891].groupby(['Sex','Pclass','Title']).median().Age
df_combine_train = df_combine_train.reset_index(name = 'Age')
```

```python
def find_age(row):
    condition = ((row['Sex'] == df_combine_train['Sex'])&(row['Pclass'] == df_combine_train['Pclass'])&\
                 (row['Title'] == df_combine_train['Title']))
    return df_combine_train[condition]['Age'].values[0]

def fill_age():
    df_combine['Age'] = df_combine.apply(lambda row:find_age(row) \
                                                    if np.isnan(row['Age']) else row['Age'], axis = 1)
    return df_combine

df_combine = fill_age()
df_combine.drop('Title', axis=1, inplace=True)
```

### Process embarked 

The Embarked feature is changed to dummy variables with prefix "Embarked". The original column is then dropped.

```python
def process_embarked():
    global df_combine
    # two missing embarked values - filling them with the most frequent one in the train  set(S)
    df_combine.Embarked.fillna('S', inplace=True)

    embarked_dummies = pd.get_dummies(df_combine['Embarked'], prefix='Embarked')
    df_combine = pd.concat([df_combine, embarked_dummies], axis=1)
    df_combine.drop('Embarked', axis=1, inplace=True)

    return df_combine
```

```python
df_combine = process_embarked()
```

```python
df_combine.head()
```

### Process Cabin

The Cabin is changed to dummy variables.

```python
def process_cabin():
    global df_combine
    df_combine.Cabin.fillna('N', inplace = True)
    df_combine['Cabin'] = df_combine['Cabin'].apply(lambda row:row[0])
    return df_combine
```

```python
df_combine = process_cabin()
```

```python
cabin_dummies = pd.get_dummies(df_combine['Cabin'], prefix = 'Cabin')
df_combine = pd.concat([df_combine, cabin_dummies], axis = 1)
df_combine.drop('Cabin', axis = 1, inplace = True)
```

```python
df_combine.head()
```

### Process Sex

Male is denoted as 0 while female is denoted as 1.

```python
#male is 0 and female is 1
def process_sex():
    global df_combine
    df_combine['Sex'] = df_combine['Sex'].map({'male':0, 'female':1})
    return df_combine    
```

```python
df_combine = process_sex()
```

### Process Pclass

Pclass is changed to dummy variables.

```python
def process_pclass():
    global df_combine
    Pclass_dummies = pd.get_dummies(df_combine['Pclass'], prefix = 'Pclass')
    df_combine = pd.concat([df_combine, Pclass_dummies], axis = 1)
    df_combine.drop('Pclass', axis = 1,inplace = True)
    return df_combine
```

```python
df_combine = process_pclass()
```

### Process Ticket

The ticket number is stripped from the dataset. Only the ticket letters are kept. Ticket is then changed to dummy variables.

```python
df_combine['Ticket'] = df_combine['Ticket'].apply(lambda x:x.replace('.',''))
df_combine['Ticket'] = df_combine['Ticket'].apply(lambda x:x.replace('/',''))
```

```python
df_combine['Ticket'] = df_combine['Ticket'].str.extract('([a-zA-Z ]+)', expand=False).str.strip()
df_combine['Ticket'].fillna('XXX', inplace = True)
```

```python
def process_ticket():
    global df_combine
    ticket_dummies = pd.get_dummies(df_combine['Ticket'], prefix = 'Ticket')
    df_combine = pd.concat([df_combine, ticket_dummies], axis = 1)
    df_combine.drop('Ticket', axis = 1, inplace = True)
    return df_combine
```

```python
df_combine = process_ticket()
```

```python
df_combine.head()
```

### Process Family

A new column "Family" is created by addition of the sibsp and Parch wich denotes the size of the family. And then the famliy is categorized into 3 new categories base on the size of the family. 

```python
df_combine['Family'] = df_combine['SibSp'] + df_combine['Parch']
```

```python
def process_family():
    df_combine['Family_Alone'] = df_combine['Family'].map(lambda x: 1 if x == 0 else 0)
    df_combine['Family_Small'] = df_combine['Family'].map(lambda x: 1 if (x >=1) & (x <=3) else 0)
    df_combine['Family_Big'] = df_combine['Family'].map(lambda x: 1 if x >=4 else 0)
    df_combine.drop(['SibSp', 'Parch'], axis = 1, inplace = True)
    return df_combine
```

```python
df_combine = process_family()
```

```python
df_combine.head()
```

### Process Fare

```python
df_combine.Fare.fillna(df_combine.Fare.mean, inplace = True)
```
